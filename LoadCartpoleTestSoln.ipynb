```python
from CartpoleProblem import build_model, getEnv
import numpy as np
import time
```

```python
model = build_model()
```

```python
#Weights not loaded, model will run but do poorly
env = getEnv()

done = False
obs = env.reset()
obs = np.array([obs])
cumReward = 0
while not done:
    print(obs)
    env.render()
    action = np.argmax(model.predict(obs)[0])
    obs[0], reward, done, _ = env.step(action)
    cumReward += reward
    time.sleep(0.01)
print(cumReward)
```
