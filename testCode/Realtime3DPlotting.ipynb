```python
#2D example
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import time

n = 10
matrix = np.random.normal(0,1,2*n).reshape(n,2)


fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,n):
    ax.clear()
    ax.set_ylim([-3,3])
    ax.set_xlim([-3,3])

    ax.plot(matrix[0:i,0], matrix[0:i,1], 'ro')
    fig.canvas.draw()
    time.sleep(0.05)
    
print("hello!")
```

```python
#3D example
n = 100
matrix = np.random.normal(0,1,3*n).reshape(n,3)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.ion()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.ion()
fig.show()
fig.canvas.draw()

for i in range(0,n):
    ax.clear()
    ax.set_xlim([-3,3])
    ax.set_ylim([-3,3])
    ax.set_zlim([-3,3])
    ax.plot(matrix[0:i,0], matrix[0:i,1], matrix[0:i,2], 'ro')
    fig.canvas.draw()
    time.sleep(0.05)
plt.ioff()
```
